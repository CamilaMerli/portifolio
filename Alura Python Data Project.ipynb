```python

# Atribuir a coluna com nomenclatura da CDU
def categoriza_cdu(cdu_numero):
  if(cdu_numero < 100):
    return 'Generalidades'
  elif(cdu_numero < 200):
    return 'Filosofia e psicologia'
  elif(cdu_numero < 300):
    return 'Religião'
  elif(cdu_numero < 400):
    return 'Ciências sociais'
  elif(cdu_numero < 500):
    return 'Classe vaga'
  elif(cdu_numero < 600):
    return 'Matemática e ciências naturais'
  elif(cdu_numero < 700):
    return 'Ciências aplicadas'
  elif(cdu_numero < 800):
    return 'Belas artes'
  elif(cdu_numero < 900):
    return 'Linguagem'
  else:
    return 'Geografia. Biografia. História.'

#  essa função a cada valor da coluna e criamos a nova coluna de uma vez
emprestimos_completo['CDU_geral'] = emprestimos_completo['localizacao'].apply(categoriza_cdu)

emprestimos_completo.head()
```

```python
#Excluir coluna de registro
emprestimos_completo.drop(columns=['registro_sistema'],inplace=True)


#modificar o tipo de dado da matrícula

emprestimos_completo['matricula_ou_siape'] = emprestimos_completo['matricula_ou_siape'].astype('string')
```
